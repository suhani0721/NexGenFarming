Taking only Common crops into consideration


```python
import pandas as pd
df1 = pd.read_csv("../Datasets/crop_production.csv")
df2 = pd.read_csv("../Datasets/crop_recommendation.csv")
df1["Crop_clean"] = df1["Crop"].str.strip().str.lower()
df2["label_clean"] = df2["label"].str.strip().str.lower()
common_crops = set(df2["label_clean"])
final_df = df1[df1["Crop_clean"].isin(common_crops)]
final_df = final_df.drop(columns=["Crop_clean"])
final_df.to_csv("filtered_file1.csv", index=False)

print(final_df.head())
```
