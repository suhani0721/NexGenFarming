```python
import requests

def get_weather(latitude, longitude):
    url = "https://api.open-meteo.com/v1/forecast"
    
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "current_weather": True
    }

    response = requests.get(url, params=params)
    data = response.json() 

    return data["current_weather"]

# Example: Hyderabad
weather = get_weather(17.3850, 78.4867)
print(weather)
```
